```python
import numpy as np
from sklearn import tree

DATA_FILES = [["/Users/jan-erikhuhne/Documents/NISE/NeuroInspiredSystemEngineering-/data/zero.txt",0],
              ["/Users/jan-erikhuhne/Documents/NISE/NeuroInspiredSystemEngineering-/data/one.txt",1],
              ["/Users/jan-erikhuhne/Documents/NISE/NeuroInspiredSystemEngineering-/data/two.txt",2],
              ["/Users/jan-erikhuhne/Documents/NISE/NeuroInspiredSystemEngineering-/data/three.txt",3],
              ["/Users/jan-erikhuhne/Documents/NISE/NeuroInspiredSystemEngineering-/data/four.txt",4],
              ["/Users/jan-erikhuhne/Documents/NISE/NeuroInspiredSystemEngineering-/data/five.txt",5],
              ["/Users/jan-erikhuhne/Documents/NISE/NeuroInspiredSystemEngineering-/data/six.txt",6],
              ["/Users/jan-erikhuhne/Documents/NISE/NeuroInspiredSystemEngineering-/data/seven.txt",7],
              ["/Users/jan-erikhuhne/Documents/NISE/NeuroInspiredSystemEngineering-/data/eight.txt",8],
              ["/Users/jan-erikhuhne/Documents/NISE/NeuroInspiredSystemEngineering-/data/nine.txt",9]
              ]


x = np.empty(shape=[0, 11])

y = np.array([])

for file, cl in DATA_FILES:
    i = 0
    with open(file, "r") as f:
        lines = f.readlines()
        for sample in lines:
            if sample:
                sample_data = np.array([float(s) for s in sample.split(" ")[:-1]])[np.newaxis, :]
                x = np.append(x, sample_data, axis=0)
                i += 1
    for _ in range(i):
        y = np.append(y, cl)
   
print(x.shape)
print(y.shape)


clf = tree.DecisionTreeClassifier()

clf = clf.fit(x, y)
```

```python
tree.plot_tree(clf)
```
