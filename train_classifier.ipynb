```python
import numpy as np
from sklearn import tree

DATA_FILES = [
    [r"C:/Users/m_keh/NeuroInspiredSystemEngineering-/data4/zero.txt", 0],
    [r"C:/Users/m_keh/NeuroInspiredSystemEngineering-/data4/one.txt", 1],
    [r"C:/Users/m_keh/NeuroInspiredSystemEngineering-/data4/two.txt", 2],
    [r"C:/Users/m_keh/NeuroInspiredSystemEngineering-/data4/three.txt", 3],
    [r"C:/Users/m_keh/NeuroInspiredSystemEngineering-/data4/four.txt", 4],
    [r"C:/Users/m_keh/NeuroInspiredSystemEngineering-/data4/five.txt", 5],
    [r"C:/Users/m_keh/NeuroInspiredSystemEngineering-/data4/six.txt", 6],
    [r"C:/Users/m_keh/NeuroInspiredSystemEngineering-/data4/seven.txt", 7],
    [r"C:/Users/m_keh/NeuroInspiredSystemEngineering-/data4/eight.txt", 8],
    [r"C:/Users/m_keh/NeuroInspiredSystemEngineering-/data4/nine.txt", 9],
    [r"C:/Users/m_keh/NeuroInspiredSystemEngineering-/data4/RnR.txt", 10],
    [r"C:/Users/m_keh/NeuroInspiredSystemEngineering-/data4/thumbsup.txt", 11]
]



x = np.empty(shape=[0, 15])

y = np.array([])

for file, cl in DATA_FILES:
    i = 0
    with open(file, "r") as f:
        lines = f.readlines()
        for sample in lines:
            if sample:
                sample_data = np.array([float(s) for s in sample.split(" ")[:-1]])[np.newaxis, :]
                x = np.append(x, sample_data, axis=0)
                i += 1
    for _ in range(i):
        y = np.append(y, cl)
   
print(x.shape)
print(y.shape)

print(y)
clf = tree.DecisionTreeClassifier()

clf = clf.fit(x, y)
```

```python
tree.plot_tree(clf)
```
